# Statistical Analysis

Column 'num_pages' has some non numerical values which we have to either drop or replace it by mean value of the column 'num_pages'.
For which I have accessed the 'num_pages' column and sorted it with 10 tail values which will show the values which are non numeric i.e eng,en-US. After this thid i have dropped that values from the column. Afterwards calculated the mean of the column. Finally replaced the mean value of accessed column in place of 'eng, eng-US' values in the main dataset.
Here onwards I have used this cleaned dataset for operations.

```python
# num_pages column having non numerical values
access=data.sort_values(by="  num_pages", ascending=True).tail(10)
access
```

```python
# dropping un-neccesory values
access.drop(access.loc[access['  num_pages']=='eng'].index, inplace=True)
access.drop(access.loc[access['  num_pages']=='en-US'].index, inplace=True)
access
```

```python
# calculating mean for filling at place of un 
mean=access['  num_pages'].mean()
mean
```

```python
# replacing eng and en-US values with mean value of column

data['  num_pages'] = data['  num_pages'].replace(['eng'],'1.6549')
data['  num_pages'] = data['  num_pages'].replace(['en-US'],'1.6549')
data['  num_pages'].sort_values(ascending=True)
```

```python
# printing dataset again 
data
```

```python
#  arranging the num_pages column in descending order
data.sort_values(by="  num_pages", ascending=False)
```

Q.1 What is the most popular book?

```python
p=data.sort_values(by="average_rating", ascending=True).tail(30)
p
```

 In the plot below , we will see that the books rated 5 will be the most popular book as the top rating is 5

```python
avg_ratings=p['average_rating']
title=p['title']
plt.rcParams["figure.figsize"] = [8, 9]
plt.bar(avg_ratings, title,color='red')
plt.title('MOST POPULAR BOOK ')
plt.xlabel('AVerage ratings')
plt.ylabel('Title of books')
plt.show()
```

Q.2 Are books with fewer pages rated higher than those with large page counts ?

```python
df=data[['  num_pages','average_rating']]
df
```

```python
sort=df.sort_values(by="average_rating", ascending=True).tail(30)
lim=sort[0:10]
lim
```

```python
a=lim['  num_pages']
b=lim['average_rating']
plt.rcParams["figure.figsize"] = [10, 6]
plt.bar(a[0:20],b[0:20],color='black')
plt.title('most rated book with fewest no. of pages ')
plt.xlabel('No of pages')
plt.ylabel('AVerage ratings')
plt.show()
```

     In the plot above , we notice that rating of book is irrespective of the no of pages of the book . From above graph it is clear that  the book with fewest no of page i.e 56 pages has the most rating i.e 5 .

#Q.7 What is Houghton Mifflin Harcourt's most popular book?

```python
sub=data[['publisher','title','average_rating']]
sub
```

```python
Q=(sub['publisher']=='Houghton Mifflin Harcourt')
Q
```

```python
P=sub[Q]
P
```

```python
x=P.sort_values(by='average_rating',ascending=True).tail(10)
x
```

```python
books=x['title']
rating=x['average_rating']
plt.bar(rating,books,color='cyan')
plt.rcParams["figure.figsize"] = [7, 5]
plt.title(' Houghton Mifflin Harcourts most popular book')
plt.xlabel('Average rating')
plt.ylabel('Top rated books')
plt.show()
```

**From above plot we can clearly conclude that Houghton Mifflin Harcourts most popular book is 'The J.R.R. Tolkien Companion and Guide' rated most i.e. 4.64**
